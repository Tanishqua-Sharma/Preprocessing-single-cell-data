```python
data_mat = adata_pp.X.T
# convert to CSC if possible. See https://github.com/MarioniLab/scran/issues/70
if issparse(data_mat):
    if data_mat.nnz > 2**31 - 1:
        data_mat = data_mat.tocoo()
    else:
        data_mat = data_mat.tocsc()
ro.globalenv["data_mat"] = data_mat
ro.globalenv["input_groups"] = adata_pp.obs["groups"]
```

```python
data_mat = adata_pp.X.T -- scran wants gene x cell matrix but scanpy has cell X gene
if issparse(data_mat): -- if matrix is sparse
if data_mat.nnz > 2**31 - 1: -- nnz - non zero entities , 2**31 - 2 BILLION
if greater than it switches to COO otherwise convert to CSC
```

```python
COO:
Suppose I have this matrix
0  5  0
3  0  0
0  0  8
non zero elements - 5 (0,1), 3(1,0) and 8 (1,2)
C00 stores
rows    = [0, 1, 2]
cols    = [1, 0, 2]
data    = [5, 3, 8]
COO is safer because:
It stores explicit coordinate lists
No compressed pointer structure
Less risk of overflow
```

```python
CSC:
0  5  0
3  0  0
0  0  8
non zero elements - 5 (0,1), 3(1,0) and 8 (1,2)
data    = [5, 3, 8]
indices = [1, 0, 2] #row
indptr  = [0, 1, 2, 3]
For 5 - data[0:1] → [5] , it starts at 0th row and ends before before 1 
```

```python
%%R -o size_factors

size_factors = sizeFactors(
    computeSumFactors(
        SingleCellExperiment(
            list(counts=data_mat)), 
            clusters = input_groups,
            min.mean = 0.1,
            BPPARAM = MulticoreParam()
    )
)
```

```python
Raw Count Matrix (genes × cells)
            ↓
Cluster Cells (Leiden)
            ↓
Work Within Each Cluster Separately
            ↓
Filter Low-Expression Genes (min.mean = 0.1)
            ↓
Create Overlapping Pools of Cells
            ↓
Sum Counts Within Each Pool
            ↓
Estimate Size Factors for Each Pool
            ↓
Solve Linear System (Deconvolution)
            ↓
Obtain Size Factor Per Cell
            ↓
Return Size Factors to Python
```

```python
Filter Low-Expression Gene -- min.mean = 0.1
Instead of estimating size factor per cell directly,
scran:
Creates overlapping pools like:
Pool 1 = Cell 1 + Cell 2 + Cell 3
Pool 2 = Cell 2 + Cell 3 + Cell 4
Pool 3 = Cell 3 + Cell 4 + Cell 5
For each pool:
Pool Size Factor= Total pooled counts/Expected counts
```

```python
| Gene | Cell1 | Cell2 | Cell3 | Cell4 |
| ---- | ----- | ----- | ----- | ----- |
| G1   | 100   | 200   | 300   | 400   |
| G2   | 50    | 100   | 150   | 200   |
| G3   | 20    | 40    | 60    | 80    |
Notice something:
For gene 1 - cell2/cell1 = 200/100 = 2 , Each cell is exactly 2× the previous.
True size factor should be :
cell 1 - 1, cell 2 - 2 , cell 3 - 3 ...
Step 1: Why Not Just Use Total Counts?
Cell1 = 170
Cell2 = 340
Cell3 = 510         
Cell4 = 680
That already reflects scaling.
    
But in real data:
Many genes are zero
Dropout noise exists
Cell types differ

Create Pools
Pool A = Cell1 + Cell2
Pool B = Cell2 + Cell3
Pool C = Cell3 + Cell4
    
Pool A:
| Gene | Count |
| ---- | ----- |
| G1   | 300   |
| G2   | 150   |
| G3   | 60    |

For pool B:

| Gene | Count |
| ---- | ----- |
| G1   | 500   |
| G2   | 250   |
| G3   | 100   |

For pool c:
| Gene | Count |
| ---- | ----- |
| G1   | 700   |
| G2   | 350   |
| G3   | 140   |
    

How is pool size factor calculated
SCRAN uses a median ratio method (similar to DESeq).
    
Size factor = median(pool gene count/reference)

Reference :
G1 = (300×500×700)^1/3 = 470
G2 = 235
G3 = 94

Pool A:
  
G1 = 300/470 = 0.64
G2 = 150/235 = 0.64
G3 = 60/94 = 0.64
   SFA=0.64 

Pool B : SFB=1.06
Pool c: SFC=1.49

Proportion of SFA:SFB:SFC[3 : 5 : 7] = (1+2) : (2+3) : (3+4)
SFA=s1+s2 -- s1 = 1,s2 = 2
```

```python
adata.obs["size_factors"] = size_factors #technical bias
scran = adata.X / adata.obs["size_factors"].values[:, None]
adata.layers["scran_normalization"] = csr_matrix(sc.pp.log1p(scran))
```
